```python
import pandas as pd
import numpy as np

data={
    "Age":[25,30,35,40,45,21,34,23,23,45,15,12,67,89,34],
    "Salary":[50000,60000,70000,80000,90000,55000,65000,75000,75000,95000,45000,40000,100000,120000,75000],
}
```

```python
df=pd.DataFrame(data)
```

```python
from sklearn.cluster import KMeans

kmean=KMeans(n_clusters=3)

df["Cluster"]=kmean.fit_predict(df)
```

```python
import matplotlib.pyplot as plt
wcss=[]

for i in range(1,11):
    kmean=KMeans(n_clusters=i)
    kmean.fit(df)
    wcss.append(kmean.inertia_)

plt.plot(range(1,11),wcss)
```

```python
from sklearn.cluster import KMeans

kmean=KMeans(n_clusters=4)

df["Cluster"]=kmean.fit_predict(df)
```

```python
df["Cluster"]
```

```python
for i in range(len(df)):
    if df["Cluster"][i]==0:
        plt.scatter(df["Age"][i],df["Salary"][i],color="red")
    elif df["Cluster"][i]==1:
        plt.scatter(df["Age"][i],df["Salary"][i],color="blue")
    elif df["Cluster"][i]==2:
        plt.scatter(df["Age"][i],df["Salary"][i],color="yellow")
    else:
        plt.scatter(df["Age"][i],df["Salary"][i],color="green")
```
