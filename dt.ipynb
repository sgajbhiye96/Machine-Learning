```python
import numpy as np
import seaborn as sns
```

```python
x=[1,2,3,4,5,6,50]
x=np.array(x)
```

```python
sns.distplot(x)
```

```python
sns.boxplot(x)
```

```python
upper_limit=np.quantile(x,0.95)
lower_limit=np.quantile(x,0.05) 
# np.where((x > upper_limit) | (x < lower_limit))
x=np.where((x < upper_limit) & (x > lower_limit))
```

```python
sns.distplot(x)
```

```python
sns.boxplot(x)
```

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.pipeline import Pipeline
```

```python
data=pd.DataFrame({
    "age":[20,21,22,23,24,25],
    "salary":[1000,2000,3000,4000,5000,6000],
    "city":["a","b","c","a","b","c"],
    "experinece":[1,2,3,4,5,6]})
```

```python
data
```

```python
ct=ColumnTransformer(transformers=[
    ("ohe",OneHotEncoder(),[2]),
    ("num",StandardScaler(),[0,1,3])
])
```

```python
ct.fit_transform(data)
```
